# FinMind
[官方說明](https://finmind.github.io/)
* 需註冊
* 有限流量 (每小時使用上限600次)
* 技術面 : 台股股價 daily、歷史 tick 逐筆交易、PER、PBR、每5秒委託成交統計、台股加權指數。
* 基本面 : 綜合損益表、現金流量表、資產負債表、股利政策表、除權除息結果表、月營收。
* 籌碼面 : 外資持股、股權分散表(2010起)、融資融券、三大法人買賣、借券成交明細。
* 消息面 : 台股相關新聞。
* 衍生性商品 : 期貨、選擇權 daily data、即時報價、交易明細。
* 國際市場 : 美股股價 daily、minute、美國債券殖利率、貨幣發行量(美國)、黃金價格、原油價格、G8 央行利率、G8 匯率、

```python
from FinMind.data import DataLoader
apiToken = "你的 API TOKEN"
api = DataLoader()
api.login_by_token(api_token=apiToken)
```

```python
df = api.taiwan_stock_holding_shares_per(
    stock_id="2330",
    start_date='2010-01-01',
    end_date='2013-04-12'
)
df
```

# FinLab
[官方說明](https://ai.finlab.tw/database)

* 需註冊
* 有回測工具
* 開高低收 (2010起)
* 很多其他數據
* 有限流量 (每日資料用量500MB)

```python
from finlab import data
close = data.get('price:收盤價')
close
```

```python
# 回測工具 https://doc.finlab.tw/getting-start/
from finlab import backtest
position = close >= close.rolling(300).max()    # position = 
report = backtest.sim(position, resample='M')
report.display()
```
